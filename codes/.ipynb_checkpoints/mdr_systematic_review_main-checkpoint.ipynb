```python
import plotly.express as px
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.gridspec as gridspec
import re
import forestplot as fp
from matplotlib import pyplot
from matplotlib.colors import LogNorm
pd.set_option("display.precision", 2)

df = pd.read_excel(r'C:\Users\Tan Wei Cong\OneDrive - National University of Singapore\NUS\Systematic Review on Resistance Definitions\data extraction\coding\drugresistance_compiled.xlsx', sheet_name='data')

df1=df.query("included in (1,0,'duplicate')")
df_inc = df1[df1['included']==1]
df_exc = df1[df1['included']==0]

df_inc = df_inc.fillna('NA') # for "other_organism" and "comments" columns
df_inc = df_inc.replace(['skin and soft tissue'],'others')
df_inc = df_inc.replace(['brain and meningeal'],'others')
df_inc = df_inc.replace(['surgical site infection'],'others')
df_inc = df_inc.replace(['intraabdominal'],'gi or intraabdominal')

gastrolist = df_inc[df_inc['comment'].str.contains('GI|gastro|enteritis|Gastro|Typhoid|typhoid|typhi|Typhi|salmonella|Salmonella|Shigella|shigella|shigellosis|Shigellosis|diarrhea|Diarrhea|diarrhoea|Diarrhoea|Helicobacter|helicobacter|campylobacter|Campylobacter')]
for i in list(gastrolist.index):
    if df_inc.loc[i]['infection_type'] == 'others':
        df_inc.loc[i,'infection_type'] = 'gi or intraabdominal'
    else:
        continue
       
df_inc = df_inc.astype({'year': 'int64',
                        'mdr_definition': 'int64','xdr_definition': 'int64','pdr_definition': 'int64','dtr_definition': 'int64','udr_definition': 'int64',
                        'usedinmethod': 'int64','usedinresult': 'int64'})
print(df_inc.head(5))
print(df1['included'].value_counts())
print(df1['included'].shape[0])
```

```python
df_analysis = pd.DataFrame()
df_analysis['number'] = df_inc['number']
df_analysis['title'] = df_inc['title']
df_analysis['year'] = df_inc['year']
df_analysis['study_type'] = df_inc['study_type']
df_analysis['infection'] = df_inc['infection_type'] 
df_analysis['infection'] = df_analysis['infection'].replace(['NA'],'non-clinical')
df_analysis['infection'] = df_analysis['infection'].replace(['others'],'other infections')

studypop=[]
for x in df_inc['study_population']:
    if '-' in x:
        studypop.append('Mixed Population')
    elif x == 'unknown' or x == 'NA':
        studypop.append('Not specified/described at all') #same as missing actually
    else:
        studypop.append(x)
df_analysis['studypop'] = studypop

term = []
for x in list(zip(df_inc['mdr_definition'],df_inc['xdr_definition'],df_inc['pdr_definition'])):
    if x == (1,0,0):
        term.append('mdr only')
    elif x == (1,1,1):
        term.append('mdr, xdr and pdr')
    else:
        term.append('mdr, xdr and pdr in various combinations')
df_analysis['term'] = term

typing = []
for x in list(zip(df_inc['genotype_resistance'],df_inc['phenotype_resistance'])):
    if x == (0,1):
        typing.append('phenotype only')
    elif x == (1,1):
        typing.append('phenotype and genotype')
    elif x == (1,0):
        typing.append('genotype only')
    else:
        typing.append('no typing done/specified') #not considered missing; column has no NA
df_analysis['typing'] = typing

susguide = []
for x in list(zip(df_inc['eucast_susceptguide'],df_inc['clsi_susceptguide'],df_inc['others_susceptguide'])):
    if x == (1,0,0):
        susguide.append('eucast only')
    elif x == (1,1,0):
        susguide.append('eucast and clsi')
    elif x == (0,1,0):
        susguide.append('clsi only')
    elif x == (0,0,1):
        susguide.append('other guidelines only')
    else:
        susguide.append('no guidelines used/specified') #phenotyping not done + phenotyping done but guidelines not specified
df_analysis['susguide'] = susguide

paeruindex = df_inc[df_inc['other_organism_specify'].str.contains('Pseudomonas|pseudomonas')].index
df_inc.loc[paeruindex,'paeru_organism'] = 1.0
abaumindex = df_inc[df_inc['other_organism_specify'].str.contains('Acinetobacter|acinetobacter')].index
df_inc.loc[abaumindex,'abaum_organism'] = 1.0
microb = []

entero = []
for x in df_inc['enterobacterales_organism']:
    if x == 1.0:
        entero.append(1)
    else:
        entero.append(0)
df_analysis['entero'] = entero

nonferm = []
for x in list(zip(df_inc['paeru_organism'],df_inc['abaum_organism'])):
    if x == (1,0) or x == (0,1) or x == (1,1):
        nonferm.append(1)
    else:
        nonferm.append(0)
df_analysis['nonfermenters'] = nonferm

grampos1 = []
for x in list(zip(df_inc['saureus_organism'],df_inc['enterococcus_organism'])):
    if x == (0,0):
        grampos1.append(0)
    else:
        grampos1.append(1)
df_inc['grampos1'] = grampos1

grampospattern = pd.read_excel(r'C:\Users\Tan Wei Cong\OneDrive - National University of Singapore\NUS\Systematic Review on Resistance Definitions\data extraction\coding\gramposlist.xlsx',sheet_name='gramposlist')
# grampospattern = pd.read_excel(r'C:\Users\Tan_W\OneDrive - National University of Singapore\NUS\Systematic Review on Resistance Definitions\data extraction\coding\gramposlist.xlsx',sheet_name='gramposlist')
grampos2 = []
otherneg = []
for x in df_inc['other_organism_specify']:
    if x == 'NA': #if that row has no string - other_organism is 0
        grampos2.append(0)
        otherneg.append(0)
    elif len(x.split(',')) == 1: #if that row has only one string (aka one microbe)
        temp = [] #temp list to see if that string matches any of the grampospatterns
        for ptn in grampospattern['pattern']:
            match = re.search(ptn,x,re.IGNORECASE)
            if match is None: #if in that row,
                temp.append(0)      
            else:
                temp.append(1)
        if 1 in temp: #as long as there's one match, means there's gram positive
            grampos2.append(1)
            otherneg.append(0)
        else:
            grampos2.append(0)
            otherneg.append(1)
    else:
        temp1 = [] #temp list to see if those strings match any of the grampospatterns, yes means there's gram pos and gram neg, no means only gram neg
        for i in x.split(','):
            temp2 = [] #temp list to see if that string OF STRINGS in that row matches any of the grampospatterns
            for ptn in grampospattern['pattern']:
                match = re.search(ptn,i,re.IGNORECASE)
                if match is None: #if in that row,
                    temp2.append(0)
                else:
                    temp2.append(1)
            if 1 in temp2:
                temp1.append(1)
            else:
                temp1.append(0)
        if 1 in temp1 and 0 in temp1:
            grampos2.append(1)
            otherneg.append(1)
        elif 1 in temp1 and 0 not in temp1:
            grampos2.append(1)
            otherneg.append(0)
        else:
            grampos2.append(0)
            otherneg.append(1)
df_inc['grampos2'] = grampos2
df_analysis['otherneg'] = otherneg

grampos=[]
for x in list(zip(df_inc['grampos1'],df_inc['grampos2'])):
    if x == (0,0):
        grampos.append(0)
    else:
        grampos.append(1)
df_analysis['grampos'] = grampos

cite = []
for row in df_inc['cite_definition']:
    if row == 'cited Magiorakos 2011':
        cite.append(1)
    else:
        cite.append(0)
df_analysis['cite'] = cite

incometier = []
income = []
incomedf = pd.read_excel(r'C:\Users\Tan Wei Cong\OneDrive - National University of Singapore\NUS\Systematic Review on Resistance Definitions\data extraction\coding\CLASS_incomelevel.xlsx', sheet_name='List of economies')
# incomedf = pd.read_excel(r'C:\Users\Tan_W\OneDrive - National University of Singapore\NUS\Systematic Review on Resistance Definitions\data extraction\coding\CLASS_incomelevel.xlsx', sheet_name='List of economies')
for row in df_inc['country']: #for each article
    if ',' in row:
        incometier.append(5)
    else:
        if row in incomedf['economy'].unique(): #if found under economy, check income_tier
            temp = incomedf['income_tier'][incomedf[incomedf['economy']== row].index.values].values[0]
            incometier.append(temp)
        else:
            incometier.append(0)

for x in incometier:
    if x == 4:
        income.append('High income')
    elif x == 3:
        income.append('Upper middle income')
    elif x == 2:
        income.append('Lower middle income')
    elif x == 1:
        income.append('Lower income')
    elif x == 5:
        income.append('Mixed income groups')
    else:
        income.append('Not found')
df_analysis['income'] = income

# print(df_analysis.head(5))
df_analysis
```

```python
#Create univariate logistic regression with all these data in a table
modeldata = df_analysis4[['cite','study_type','infection','term','typing','susguide','entero', 'nonfermenters', 'grampos','otherneg']]

#study type
model_study = smf.logit("cite ~ C(study_type, Treatment(reference='observational'))",data=modeldata[['cite','study_type']]).fit()

params_study = model_study.params
print(model_study.summary())

conf_study = model_study.conf_int() #confidence interval
conf_study['Odds Ratio'] = params_study
conf_study.columns = ['2.5%', '97.5%', 'Odds Ratio']

# convert log odds3 to ORs
odds3_study = pd.DataFrame({"Odds Ratio": conf_study['Odds Ratio'],
                            "Lower CI": conf_study['2.5%'],
                            "Upper CI": conf_study['97.5%']})

odds3_study = np.exp(odds3_study)

# check if pvalues are significant
odds3_study['P-values'] = model_study.pvalues
# odds3_study['Significant?'] = ['Significant' if pval <= 0.05 else 'Not significant' for pval in model_study.pvalues]


#infection
model_infect = smf.logit("cite ~ C(infection, Treatment(reference='screening samples'))",data=modeldata[['cite','infection']]).fit()

params_infect = model_infect.params
print(model_infect.summary())

conf_infect = model_infect.conf_int() #confidence interval
conf_infect['Odds Ratio'] = params_infect
conf_infect.columns = ['2.5%', '97.5%', 'Odds Ratio']

# convert log odds3 to ORs
odds3_infect = pd.DataFrame({"Odds Ratio": conf_infect['Odds Ratio'],
                            "Lower CI": conf_infect['2.5%'],
                            "Upper CI": conf_infect['97.5%']})

odds3_infect = np.exp(odds3_infect)

# check if pvalues are significant
odds3_infect['P-values'] = model_infect.pvalues
odds3_infect['Significant?'] = ['Significant' if pval <= 0.05 else 'Not significant' for pval in model_infect.pvalues]

#terminology
model_term = smf.logit("cite ~ C(term, Treatment(reference='mdr only'))",data=modeldata[['cite','term']]).fit()

params_term = model_term.params
print(model_term.summary())

conf_term = model_term.conf_int() #confidence interval
conf_term['Odds Ratio'] = params_term
conf_term.columns = ['2.5%', '97.5%', 'Odds Ratio']

# convert log odds3 to ORs
odds3_term = pd.DataFrame({"Odds Ratio": conf_term['Odds Ratio'],
                            "Lower CI": conf_term['2.5%'],
                            "Upper CI": conf_term['97.5%']})

odds3_term = np.exp(odds3_term)

# check if pvalues are significant
odds3_term['P-values'] = model_term.pvalues
odds3_term['Significant?'] = ['Significant' if pval <= 0.05 else 'Not significant' for pval in model_term.pvalues]

#typing
model_typing = smf.logit("cite ~ C(typing, Treatment(reference='phenotype only'))",data=modeldata[['cite','typing']]).fit()

params_typing = model_typing.params
print(model_typing.summary())

conf_typing = model_typing.conf_int() #confidence interval
conf_typing['Odds Ratio'] = params_typing
conf_typing.columns = ['2.5%', '97.5%', 'Odds Ratio'] #95% confidence interval -> median is 50%, need to take 2.5-97.5%

odds3_typing = pd.DataFrame({"Odds Ratio": conf_typing['Odds Ratio'],
                            "Lower CI": conf_typing['2.5%'],
                            "Upper CI": conf_typing['97.5%']})

odds3_typing = np.exp(odds3_typing)

odds3_typing['P-values'] = model_typing.pvalues
odds3_typing['Significant?'] = ['Significant' if pval <= 0.05 else 'Not significant' for pval in model_typing.pvalues]


#susceptibility guidelines
model_sus = smf.logit("cite ~ C(susguide, Treatment(reference='clsi only'))",data=modeldata[['cite','susguide']]).fit() 

params_sus = model_sus.params
print(model_sus.summary())

conf_sus = model_sus.conf_int() #confidence interval
conf_sus['Odds Ratio'] = params_sus
conf_sus.columns = ['2.5%', '97.5%', 'Odds Ratio']

odds3_sus = pd.DataFrame({"Odds Ratio": conf_sus['Odds Ratio'],
                            "Lower CI": conf_sus['2.5%'],
                            "Upper CI": conf_sus['97.5%']})

odds3_sus = np.exp(odds3_sus)

odds3_sus['P-values'] = model_sus.pvalues
odds3_sus['Significant?'] = ['Significant' if pval <= 0.05 else 'Not significant' for pval in model_sus.pvalues]

#entero
model_e = smf.logit("cite ~ C(entero, Treatment(reference=0))",data=modeldata[['cite','entero']]).fit()

params_e = model_e.params
print(model_e.summary(xname = ['Intercept','Enterobacterales']))

conf_e = model_e.conf_int() #confidence interval
conf_e['Odds Ratio'] = params_e
conf_e.columns = ['2.5%', '97.5%', 'Odds Ratio']

odds3_e = pd.DataFrame({"Odds Ratio": conf_e['Odds Ratio'],
                            "Lower CI": conf_e['2.5%'],
                            "Upper CI": conf_e['97.5%']})

odds3_e = np.exp(odds3_e)
odds3_e['P-values'] = model_e.pvalues
odds3_e['Significant?'] = ['Significant' if pval <= 0.05 else 'Not significant' for pval in model_e.pvalues]
# odds3_e['label'] = 'Enterobacterales'
# odds3_e['group'] = 'Enterobacterales'
odds3_e = odds3_e.rename(index={'C(entero, Treatment(reference=0))[T.1]': 'C(entero, Treatment(reference=0))[T.Enterobacterales]'})

#nonferm
model_nonferm = smf.logit("cite ~ C(nonfermenters, Treatment(reference=0))",data=modeldata[['cite','nonfermenters']]).fit()

params_nonferm = model_nonferm.params
print(model_nonferm.summary(xname = ['Intercept','Non-fermenters']))

conf_nonferm = model_nonferm.conf_int() #confidence interval
conf_nonferm['Odds Ratio'] = params_nonferm
conf_nonferm.columns = ['2.5%', '97.5%', 'Odds Ratio']

odds3_nonferm = pd.DataFrame({"Odds Ratio": conf_nonferm['Odds Ratio'],
                            "Lower CI": conf_nonferm['2.5%'],
                            "Upper CI": conf_nonferm['97.5%']})

odds3_nonferm = np.exp(odds3_nonferm)
odds3_nonferm['P-values'] = model_nonferm.pvalues
odds3_nonferm['Significant?'] = ['Significant' if pval <= 0.05 else 'Not significant' for pval in model_nonferm.pvalues]
# odds3_nonferm['label'] = 'Non-fermenting Gram-Negatives'
# odds3_nonferm['group'] = 'Non-fermenting Gram-Negatives'
odds3_nonferm = odds3_nonferm.rename(index={'C(nonfermenters, Treatment(reference=0))[T.1]': 'C(entero, Treatment(reference=0))[T.Non-fermenting Gram-Negatives]'})

#grampos
model_grampos = smf.logit("cite ~ C(grampos, Treatment(reference=0))",data=modeldata[['cite','grampos',]]).fit()

params_grampos = model_grampos.params
print(model_grampos.summary(xname = ['Intercept','Gram Positives']))

conf_grampos = model_grampos.conf_int() #confidence interval
conf_grampos['Odds Ratio'] = params_grampos
conf_grampos.columns = ['2.5%', '97.5%', 'Odds Ratio']

odds3_grampos = pd.DataFrame({"Odds Ratio": conf_grampos['Odds Ratio'],
                            "Lower CI": conf_grampos['2.5%'],
                            "Upper CI": conf_grampos['97.5%']})

odds3_grampos = np.exp(odds3_grampos)
odds3_grampos['P-values'] = model_grampos.pvalues
odds3_grampos['Significant?'] = ['Significant' if pval <= 0.05 else 'Not significant' for pval in model_grampos.pvalues]
# odds3_grampos['label'] = 'Gram-Positives'
# odds3_grampos['group'] = 'Gram-Positives'
odds3_grampos = odds3_grampos.rename(index={'C(grampos, Treatment(reference=0))[T.1]': 'C(grampos, Treatment(reference=0))[T.Gram-Positives]'})

#otherneg
model_otherneg = smf.logit("cite ~ C(otherneg, Treatment(reference=0))",data=modeldata[['cite','otherneg']]).fit()

params_otherneg = model_otherneg.params
print(model_otherneg.summary(xname = ['Intercept','Other Gram-Negatives']))

conf_otherneg = model_otherneg.conf_int() #confidence interval
conf_otherneg['Odds Ratio'] = params_otherneg
conf_otherneg.columns = ['2.5%', '97.5%', 'Odds Ratio']

odds3_otherneg = pd.DataFrame({"Odds Ratio": conf_otherneg['Odds Ratio'],
                            "Lower CI": conf_otherneg['2.5%'],
                            "Upper CI": conf_otherneg['97.5%']})

odds3_otherneg = np.exp(odds3_otherneg)
odds3_otherneg['P-values'] = model_otherneg.pvalues
odds3_otherneg['Significant?'] = ['Significant' if pval <= 0.05 else 'Not significant' for pval in model_otherneg.pvalues]
# odds3_otherneg['label'] = 'Other Gram-Negatives'
# odds3_otherneg['group'] = 'Other Gram-Negatives'
odds3_otherneg = odds3_otherneg.rename(index={'C(otherneg, Treatment(reference=0))[T.1]': 'C(entero, Treatment(reference=0))[T.Other Gram-Negatives]'})

odds3 = pd.concat([odds3_study, odds3_infect, odds3_term, odds3_typing, odds3_sus,odds3_e, odds3_nonferm,odds3_grampos, odds3_otherneg], axis=0)
odds3 = odds3[odds3.index != "Intercept"]
#need to add a line to replace the index labels

odds3 = odds3.rename(index=lambda s: s.split('.')[1][:-1])
odds3 = odds3.rename({'in vitro experiment':'in vitro experiment',
                    'interventional': 'Interventional',
                    'systematic review/meta-analysis':'Systematic review/meta-analysis',
                    'bloodstream':'Bloodstream',
                    'gastrointestinal tract': 'Gastrointestinal tract',
                    'other infection source':'Other infection source',
                    'respiratory tract':'Respiratory tract',
                    'urinary tract':'Urinary tract',
                    'mdr, xdr and pdr':'MDR, XDR and PDR',
                    'mdr, xdr and pdr in various combinations':'MDR, XDR and PDR in various combinations',
                    'not specified': 'Not specified',
                    'phenotype and genotype': 'Phenotype and genotype',
                    'genotype only':'Genotype only',
                    'eucast and clsi':'EUCAST and CLSI',
                    'eucast only':'EUCAST only',
                    'others/not specified':'Others/not specified',
                    'Non-fermenting Gram-Negatives':'Non-fermenting Gram-negative',
                    'Gram-Positives':'Gram-positive',
                    'Other Gram-Negatives': 'Other Gram-negative'})
odds3['label'] = odds3.index.to_list()
odds3

def grouplabel(x):
    grouping = {'Study design (Ref: Observational)':['in vitro', 'review', 'interventional'],
                'Drug resistance terms used (Ref: MDR only)':['mdr', 'xdr', 'pdr', 'dtr', 'udr', 'term'],
                'Antibiotic susceptibility standards (Ref: CLSI only)':['eucast','clsi','guideline','others'],
                'Characterisation of antibiotic susceptibilities (Ref: Phenotype only)':['typ','not specified'],
                'Source of bacteria isolates (Ref: Screening samples)':['blood','gastro', 'clinical', 'infection', 'respiratory', 'urinary'],
                'missing':['missing'],'Enterobacterales':['enterob'],'Non-fermenting Gram-negative':['non-ferm'],
                'Other Gram-negative':['neg'],'Gram-positive':['pos']}
    for g in grouping:
        for val in grouping[g]:
            if val in x.lower():
                return g
            else:
                continue
                
group=[]
for i in odds3['label']:
    group.append(grouplabel(i.lower()))
    # print(i)

odds3['group'] = group
odds3.drop(odds3[odds3['group']=='missing'].index,inplace=True) #from infection column

ncount={}
ncount.update(df_analysis4['study_type'].value_counts().to_dict())
ncount.update(df_analysis4['infection'].value_counts().to_dict())
ncount.update(df_analysis4['susguide'].value_counts().to_dict())
ncount.update(df_analysis4['typing'].value_counts().to_dict())
ncount.update(df_analysis4['term'].value_counts().to_dict())
ncount.update({'enterobacterales': df_analysis4[df_analysis4['entero']==1].shape[0]})
ncount.update({'non-fermenting gram-negative': df_analysis4[df_analysis4['nonfermenters']==1].shape[0]})
ncount.update({'gram-positive': df_analysis4[df_analysis4['grampos']==1].shape[0]})
ncount.update({'other gram-negative': df_analysis4[df_analysis4['otherneg']==1].shape[0]})
print(ncount)

n=[]
for x in odds3['label']:
    n.append(str(ncount[x.lower()]))
odds3['n'] = n
odds3['P-values']=odds3['P-values'].round(2)

from forestplot.text_utils import star_pval
from forestplot.arg_validators import check_iterables_samelen

odds3 = star_pval(dataframe=odds3, pval="P-values", starpval=True, decimal_precision=2, thresholds=(0.05,0.05), symbols=("**","**"))
odds3 = odds3.replace(['0.0**'],'0.00**')
odds3 = odds3.reindex(['Interventional','Systematic review/meta-analysis','in vitro experiment','Bloodstream','Respiratory tract',
                       'Urinary tract','Gastrointestinal tract','Other infection source','MDR, XDR and PDR', 'MDR, XDR and PDR in various combinations', 
                       'Genotype only','Phenotype and genotype','Not specified','EUCAST only','EUCAST and CLSI','Others/not specified','Enterobacterales',
                       'Non-fermenting Gram-negative','Other Gram-negative','Gram-positive'])
odds3['label'] = odds3['label'].replace(['Enterobacterales'],'Enterobacterales studied')
odds3['label'] = odds3['label'].replace(['Non-fermenting Gram-negative'],'Non-fermenting Gram-negative studied')
odds3['label'] = odds3['label'].replace(['Other Gram-negative'],'Other Gram-negative studied')
odds3['label'] = odds3['label'].replace(['Gram-positive'],'Gram-positive studied')
odds3 = odds3.replace(['in vitro experiment'],'$\it{in}$ $\it{vitro}$ experiment')
odds3
# odds3.to_csv('univariatelogOR.csv', index = True)
# print(odds3.index.values.tolist())
```
